```python
import pandas as pd
import os
```

# Exercise 1

Download all experimental data in one DataFrame. 

Add 'Exp id' column to number all experiments (5 model results from one subfolder)

```python
models = 5
list1=os.listdir(os.path.join(os.getcwd(),'SIM'))
excount = 0
dflist=[]

for dir1 in list1:
    list2=os.listdir(os.path.join(os.getcwd(),'SIM',dir1))

    exlist = []
    for dir2 in list2:
        data = []

        for i in range(models):
            temp = pd.read_csv(os.path.join(os.getcwd(),'SIM',dir1,dir2,"Solution"+str(i)+".csv"))
            data.append(temp)

        if not data==[]:
            ex = pd.concat(data, axis=0, ignore_index=True) # dataframe for one exp
            ex['Exp id'] = excount
            excount += 1
            exlist.append(ex)
            if "_05_" in dir1:
                ex['Species no'] = 5
            if "_10_" in dir1:
                ex['Species no'] = 10
            if "_15_" in dir1:
                ex['Species no'] = 15
            if "_20_" in dir1:
                ex['Species no'] = 20
            if "_25_" in dir1:
                ex['Species no'] = 25
            if "_30_" in dir1:
                ex['Species no'] = 30
            if "_35_" in dir1:
                ex['Species no'] = 35
            if "_40_" in dir1:
                ex['Species no'] = 40
            if "_45_" in dir1:
                ex['Species no'] = 45
            if "_50_" in dir1:
                ex['Species no'] = 50
            
        
    if not exlist==[]:
        df = pd.concat(exlist, axis=0, ignore_index=True) # dataframe for all experiment with the same simulation settings
        dflist.append(df)

dd = pd.concat(dflist, axis=0, ignore_index=True)  #total dataframe
dd
```

# Exercise 2

Obtain data only for IP0 and IP2 models.

```python
da = dd.loc[(dd['Model']=='IP0') | ((dd['Model']=='IP2'))]
da
```

# Exercise 3

For data from exercise 2, remove all data where 'Objective Function' for 'IP0' and 'IP2' matches.

```python
del_exp_id_with_one_value = da.groupby('Exp id')['Objective Function'].agg(lambda x: len(set(x))==1)
del_exp_id_with_one_value = pd.DataFrame(del_exp_id_with_one_value)
del_exp_id_with_one_value = del_exp_id_with_one_value.reset_index()
del_exp_id_with_one_value = del_exp_id_with_one_value.rename(columns={"Objective Function": "del"})

da = pd.merge(da, del_exp_id_with_one_value, how='left', left_on='Exp id', right_on='Exp id')
da = da.loc[da['del']==False]
da
```

# Exercise 4

Find the greatest mismatch in results for model 'IP0' and 'IP2' (use data from Ex 3)


```python
mismatch = da.groupby('Exp id')['Objective Function'].agg(lambda x: abs(list(x)[0]-list(x)[1]))
mismatch = pd.DataFrame(mismatch)
mismatch = mismatch.reset_index()
mismatch = mismatch.rename(columns={"Objective Function": "mismatch"})
mismatch
```

```python
max_mismatch = mismatch['mismatch'].max()
max_mismatch
```

```python
exp_id_max_mismatch = mismatch.loc[mismatch['mismatch'].argmax(),'Exp id']
exp_id_max_mismatch
```

```python
da.loc[da['Exp id']==exp_id_max_mismatch]
```
